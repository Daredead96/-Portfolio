```python
def begin_data(name):
    """
    Загружает CSV файл по указанному имени и возвращает DataFrame.
    
    Parameters:
    name (str): Имя файла (без расширения .csv)
    
    Returns:
    pandas.DataFrame: Загруженные данные
    
    Raises:
    FileNotFoundError: Если файл не существует
    """
    full_path = path + name + '.csv'
    
    # Простая проверка существования файла
    try:
        df = pd.read_csv(full_path)
    except FileNotFoundError:
        print(f"Ошибка: Файл '{full_path}' не найден")
        return None
    
    # Вывод первых 2 строк для предварительного просмотра
    print("Первые 2 строки данных:")
    display(df.head(2))
    print("\n" + "="*50)
    
    return df
```

```python
# Вытягивает название датафрейма
def get_df_name(df) -> str:
    """
    Возвращает имя датафрейма в виде строки.
    
    Функция ищет в глобальном пространстве имен переменную,
    которая ссылается на переданный датафрейм, и возвращает её имя.
    
    Параметры:
    df (pd.DataFrame): датафрейм, для которого нужно получить имя
    
    Возвращает:
    str: имя датафрейма
    
    Пример использования:
    >>> df = pd.DataFrame({'A': [1, 2, 3]})
    >>> get_df_name(df)
    'df'
    """
    # Поиск имени переменной через глобальные объекты
    name = [x for x in globals() if globals()[x] is df]
    
    # Проверка наличия найденного имени
    if not name:
        raise ValueError("Датафрейм не найден в глобальных переменных")
        
    return name[0]
```

```python
def info_mod(df):
    """
    Выводит общую информацию и описательную статистику для DataFrame.
    
    Parameters:
    df (pandas.DataFrame): DataFrame для анализа
    
    Returns:
    None
    """
    df_name = get_df_name(df)
    sp = '=' * 20
    print(f'\n{sp} ОБЩАЯ ИНФОРМАЦИЯ ДЛЯ ДАТАСЕТА {df_name} {sp}\n')
    display(df.info())
    print(f'\n{sp} ОПИСАТЕЛЬНАЯ СТАТИСТИКА {sp}\n')
    display(df.describe().T)
    print(f'\n{sp} Дубликаты {sp}\n')
    n_duplicates = df.duplicated().sum()
    print(f'Количество дублированных строк: {n_duplicates}')
    print('-'*60)
```

```python
def plot_hist(df):
    """
    Строит гистограммы для всех столбцов DataFrame, кроме столбца 'id'.
    Графики размещаются по 3 в ряд.
    
    Parameters:
    df (pandas.DataFrame): DataFrame для построения гистограмм
    
    Returns:
    None
    """
    # Получаем список столбцов, исключая 'id'
    cols = df.loc[:, df.columns != 'id'].columns.tolist()
    
    # Определяем количество строк (по 3 графика в ряду)
    n_cols = 3
    n_rows = (len(cols) + n_cols - 1) // n_cols  # округление вверх
    
    # Создаем фигуру с подзаграфиками
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 5 * n_rows))
    axes = axes.flatten()  # превращаем в 1D-массив для удобной итерации
    
    for idx, col in enumerate(cols):
        ax = axes[idx]
        sns.histplot(df, x=col, ax=ax)
        ax.set_title(f'Распределение {col}')
    
    # Скрываем пустые подграфики, если столбцов не хватает на полный ряд
    for idx in range(len(cols), len(axes)):
        axes[idx].set_visible(False)

    plt.tight_layout()
    plt.show()
```

```python
def plot_box(df):
    """
    Строит вертикальные boxplot'ы для всех столбцов DataFrame, кроме столбца 'id'.
    Графики размещаются по 3 в ряд.
    
    Parameters:
    df (pandas.DataFrame): DataFrame для построения boxplot'ов
    
    Returns:
    None
    """
    # Получаем список столбцов, исключая 'id'
    cols = df.loc[:, df.columns != 'id'].columns.tolist()
    
    # Определяем количество строк (по 3 графика в ряду)
    n_cols = 3
    n_rows = (len(cols) + n_cols - 1) // n_cols  # округление вверх
    
    # Создаем фигуру с подзаграфиками
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 5 * n_rows))
    axes = axes.flatten()  # превращаем в 1D-массив для удобной итерации
    
    for idx, col in enumerate(cols):
        ax = axes[idx]
        sns.boxplot(data=df, y=col, ax=ax, orient='v')  # vertical boxplot
        ax.set_title(f'Распределение по выбросам {col}')
        ax.set_ylabel(col)  # подпись оси Y
        ax.set_xlabel('')  # убираем подпись оси X (она не нужна для vertical boxplot)

    # Скрываем пустые подграфики, если столбцов не хватает на полный ряд
    for idx in range(len(cols), len(axes)):
        axes[idx].set_visible(False)

    plt.tight_layout()
    plt.show()
```

```python
def corr_heatmaps(df1, df2, df3):
    """
    Строит три тепловые карты корреляций Пирсона для заданных датасетов.
    
    Parameters:
    df1, df2, df3 (pd.DataFrame): датасеты для анализа
    """
    # Собираем датасеты в список
    dfs = [df1, df2, df3]
    
    # Получаем имена датасетов
    titles = [get_df_name(df) for df in dfs]
    
    
    # Создаём фигуру с тремя подграфиками (в ряд)
    fig, axes = plt.subplots(1, 3, figsize=(18, 5))
    
    # Перебираем датасеты и строим heatmap для каждого
    for i, (df, title) in enumerate(zip(dfs, titles)):
        if 'id' in df.columns:
            df = df.drop('id', axis=1)
        # Вычисляем корреляционную матрицу
        corr = df.corr()
        
        # Строим heatmap на соответствующем подграфике
        sns.heatmap(corr, ax=axes[i], annot=True, cmap='coolwarm', center=0,
                    square=True, fmt='.2f', cbar=False)
        axes[i].set_title(f'Корреляция Пирсона — {title}', fontsize=14, pad=15)
    

    plt.tight_layout()
    plt.show()
```

## Обучение

**ФУНКЦИИ**

```python
def sampler(df, target='product'):
    """
    Разделяет данные на обучающую и тестовую выборки в соотношении 75/25
    """
    X = df.drop(columns=[target])
    y = df[target]
    
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.25, random_state=RS
    )
    
    return X_train, X_test, y_train, y_test
```

```python
def catnumer (df):
    df = df.drop(['product', 'id'], axis=1)  # axis=1 — для колонок

    # Разделяем столбцы на числовые и категориальные из тренировочных данных
    numeric_features = df.select_dtypes(include=['int', 'float']).columns.tolist()
    categorical_features = df.select_dtypes(include=['object', 'category']).columns.tolist()
    return  numeric_features,categorical_features
```

```python
def best_sercher(X_train, y_train):
    """
    Поиск лучшей модели с регуляризацией через GridSearchCV
    """
    
    # Получаем препроцессор
    preprocessor = preprocessor_rb
    
    # Базовый пайплайн без указания конкретного регрессора
    pipeline = Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('regressor', None)  # Будет выбран в GridSearch
    ])
    
   
    param_grid = [
    # L2 регуляризация (Ridge)
    {
        'regressor': [Ridge()],
        'regressor__alpha': [0.001, 0.01, 0.1, 1.0, 10.0, 100.0, 1000.0],
        'regressor__fit_intercept': [True, False],
        'regressor__solver': ['svd', 'lsqr', 'sag']
    },
    # L1 регуляризация (Lasso)
    {
        'regressor': [Lasso()],
        'regressor__alpha': [0.001, 0.01, 0.1, 1.0, 10.0, 100.0, 1000.0],
        'regressor__fit_intercept': [True, False],
        'regressor__selection': ['cyclic', 'random']
    },
    # Комбинированная регуляризация (ElasticNet)
    {
        'regressor': [ElasticNet()],
        'regressor__alpha': [0.001, 0.01, 0.1, 1.0, 10.0],
        'regressor__l1_ratio': [0.1, 0.3, 0.5, 0.7, 0.9],  # баланс L1/L2
        'regressor__fit_intercept': [True, False],
        'regressor__selection': ['cyclic', 'random']
    }
]
    # GridSearch с кросс-валидацией
    grid_search = GridSearchCV(
        pipeline,
        param_grid,
        cv=5,
        scoring='neg_root_mean_squared_error',
        n_jobs=-1,
        verbose=1
    )
    
    # Обучаем модель
    grid_search.fit(X_train, y_train)
    
    # Лучшая модель
    best_model = grid_search.best_estimator_
    
    # Метрики
    y_pred = best_model.predict(X_train)
    rmse = np.sqrt(mean_squared_error(y_train, y_pred))
    r2 = r2_score(y_train, y_pred)
    
    print(f"Лучшие параметры: {grid_search.best_params_}")
    print(f"RMSE: {rmse:.2f}")
    print(f"R2: {r2:.2f}")
    
    return best_model
```

```python
def predictor(model, X_test, y_test):
    """
    Выполняет предсказания на тестовых данных и возвращает результаты
    """
    
    # Предсказания на тестовых данных
    y_pred = model.predict(X_test)
    
    # Вычисляем метрики
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    r2 = r2_score(y_test, y_pred)
    
    # Создаем датафрейм с тестовыми данными и результатами
    test_df = X_test.copy()
    test_df['product'] = y_test  # настоящие значения
    test_df['product_predict'] = y_pred  # предсказания
    test_df['rmse'] = np.sqrt((y_test - y_pred) ** 2)  # RMSE для каждой строки
    
    # Средние значения
    mean_rmse = rmse
    mean_prediction = np.mean(y_pred)
    
    # Вывод результатов
    print(f"Результаты на тестовых данных:")
    print(f"RMSE: {rmse:.2f}")
    print(f"R2: {r2:.2f}")
    print(f"Средний RMSE: {mean_rmse:.2f}")
    print(f"Среднее количество продукта: {mean_prediction:.2f}")
    
    return test_df, mean_rmse, mean_prediction
```

## Анализ прибыли

По инструкциям необходимо сделать функции для подсчета прибыыли, функцию для краткой аналитики средних запасов.

**ФУНКЦИИ**

```python
def calculate_profit(df, resulte=None):
    """
    Рассчитывает прибыль от разработки 200 лучших скважин по предсказаниям модели.
    
    Параметры:
    - df: DataFrame с колонками 'product_predict' (предсказания) и 'product' (реальные значения)
    - resulte: список для накопления результатов (для bootstrap), либо None
    
    Возвращает:
    - прибыль в рублях (если resulte=None)
    - None (если resulte задан, результат добавляется в список в млн руб)
    """
    # Выбор 200 точек с наибольшим predicted product
    top_200 = df.nlargest(FIELD_USED, 'product_predict')
    
    # Расчёт дохода от реального объёма продукта (после бурения)
    total_income = (top_200['product'] * INCOME_ONE_PROD).sum()
    
    # Расчёт прибыли (доход - бюджет)
    profit = total_income - BUDGET
    
    if resulte is None:
        return profit
    else:
        resulte.append(profit / 1_000_000)  # сохраняем в млн руб для удобства
        return None
```

```python
def profit_analyze(mean_prediction, name):
    """
    Анализирует потенциальную прибыльность региона по среднему запасу продукта.
    
    Параметры:
    mean_prediction : float — средний запас продукта в регионе (тыс. баррелей)
    name : str — название датасета/региона
    
    Глобальная переменная:
    well_product : float — точка безубыточности (тыс. баррелей)
    
    Вывод:
    Печатает анализ отклонения от точки безубыточности в % и абсолютных значениях.
    """
    print(f'Анализ дохода датасета — {name}')
    
    # Рассчитываем разницу
    diff = mean_prediction - well_product  # положительная — обгоняет, отрицательная — отстаёт
    
    if mean_prediction < well_product:
        percent_diff = (diff / well_product) * 100
        print(f'Отставание от точки безубыточности: {percent_diff:.1f}% '
              f'(−{abs(round(diff, 2))} тыс. баррелей)')
        
    elif mean_prediction == well_product:
        print('Средний запас продукта в скважине совпадает с точкой безубыточности (0%)')
        
    else:  # mean_prediction > well_product
        percent_diff = (diff / well_product) * 100
        print(f'Превышение точки безубыточности: +{percent_diff:.1f}% '
              f'(+{round(diff, 2)} тыс. баррелей)')
    
    print('=' * 90)
```

## Риски

**ФУНКЦИИ**

```python
def bootstrap_profit_analysis(df, n_bootstrap=1000):
    """
    Применяет технику Bootstrap для анализа распределения прибыли
    
    Parameters:
    df - DataFrame с колонками 'product_predict' (предсказанный продукт)
    n_bootstrap - количество бутстрап-выборок
    """
    
    # Список для хранения прибыли каждой бутстрап-выборки
    bootstrap_profits = []
    
    # Инициализация генератора случайных чисел для воспроизводимости
    rng = np.random.RandomState(RS)
    
    for i in range(n_bootstrap):
        # 1. Генерация бутстрап-выборки размером 500
        bootstrap_sample = df.sample(n=500, replace=True, random_state=rng)
        
        # 2. Используем функцию calculate_profit для расчета прибыли
        calculate_profit(bootstrap_sample, bootstrap_profits)
    
    # Преобразуем в numpy array для удобства расчетов
    bootstrap_profits = np.array(bootstrap_profits)
    
    # Расчет статистик
    mean_profit = np.mean(bootstrap_profits)
    confidence_interval = np.percentile(bootstrap_profits, [2.5, 97.5])
    risk_of_loss = np.mean(bootstrap_profits < 0) * 100  # в процентах
    
    # Вывод результатов
    print("=== РЕЗУЛЬТАТЫ BOOTSTRAP-АНАЛИЗА ===")
    print(f"Средняя прибыль: {mean_profit:.2f} млн руб.")
    print(f"95% доверительный интервал: [{confidence_interval[0]:.2f}, {confidence_interval[1]:.2f}] млн руб.")
    print(f"Риск убытков: {risk_of_loss:.1f}%")
    print(f"Количество бутстрап-выборок: {n_bootstrap}")
    
    # Построение графика распределения прибыли
    plt.figure(figsize=(12, 6))
    
    # Гистограмма с KDE
    sns.histplot(bootstrap_profits, kde=True, bins=50, alpha=0.7, color='skyblue')
    
    # Вертикальные линии для ключевых метрик
    plt.axvline(mean_profit, color='red', linestyle='--', linewidth=2, 
                label=f'Средняя прибыль: {mean_profit:.2f} млн руб.')
    plt.axvline(confidence_interval[0], color='orange', linestyle='--', linewidth=2, 
                label=f'95% ДИ: [{confidence_interval[0]:.2f}, {confidence_interval[1]:.2f}]')
    plt.axvline(confidence_interval[1], color='orange', linestyle='--', linewidth=2)
    plt.axvline(0, color='black', linestyle='-', linewidth=1, 
                label='Граница безубыточности')
    
    # Заливка области убытков
    plt.axvspan(bootstrap_profits.min(), 0, alpha=0.3, color='red', 
                label=f'Зона убытков ({risk_of_loss:.1f}%)')
    
    plt.xlabel('Прибыль (млн руб.)')
    plt.ylabel('Частота')
    plt.title('Распределение прибыли по бутстрап-выборкам')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()
    
    return bootstrap_profits
```
